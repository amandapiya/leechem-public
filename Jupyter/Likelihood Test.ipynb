```python
# works
y_observed = np.random.normal(loc = 17, scale = 1, size = 1000)
dist2 = scipy.stats.norm
p = ["loc", "sd"]

def loglikelihood_neg(param, dist):
    loglike_neg = -np.sum(dist[0].logpdf(dist[1], *param))
    print(loglike_neg)
    return loglike_neg

max_lik = minimize(loglikelihood_neg, x0=[1]*len(p), args=[dist2, y_observed], method='nelder-mead').fun
print(max_lik)
```
